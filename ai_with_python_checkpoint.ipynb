```python
# Topic 59 Handling Missing Data
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(
    "retail_sales.csv",
    parse_dates=["Date"],
    dtype={
        "Category": "category",
        "Region": "category"
    }
)
df.head()
```

```python
df.info()
```

```python
# Detect missing Values
print(df.isna().sum())
```

```python
# Fill missing age with mean
df['Quantity']=pd.to_numeric(df['Quantity'])
df['Quantity']=df['Quantity'].fillna(df['Quantity'].mode()[0])
```

```python
# fill missing value with median
df['Sales']=df['Sales'].fillna(df['Sales'].median())
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['Region'],inplace=True)
```

```python
# Remove missing values in region and category
df['Category'] = df['Category'].cat.add_categories("0").fillna("0")
df['Region'] = df['Region'].cat.add_categories("0").fillna("0")
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
import pandas as pd
# Samples DataFrames
custormers=pd.DataFrame({
    'CustomerID': [1,2,3],
    'Name':['Alina', 'Bobi','Charlie']
})
orders=pd.DataFrame({
    'OrderID':[101,102,103,104,],
    'CustomerID':[1,2,3,4],
    'Amount':[250,150,200,300]
    })
```

```python
# Inner Merge
df_inner=pd.merge(customers,orders, on='CustomerID', how='inner')
print("Inner Merge:\n", df_inner)
```

```python
# Left Merge
df_left=pd.merge(customers,orders, on='CustomerID', how='left')
print("\nLeft Merge\n", df_left)
```

```python
# Concatenate Row Examples
df_concat=pd.concat([customers,customers])
print("\nConcatenated Rows:\n",df_concat)
```

```python
# Apply Creat Revenue Category
def revenue_category(row):
    if row ["revenue"] > 500:
        return "Hight"
    elif row ["revenue"] >=200:
        return "Medium"
    else:
        return "Low"
df["revenue_category"]=df.apply(revenue_category, axis=1)
# why apply decisions on multiple columns
```
