```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load dataset
df = pd.read_csv('/kaggle/input/datasets/muhammadumer7804/housing-price-dataset/HousePricePrediction.csv')

# Handle missing values
df = df.dropna(subset=['SalePrice'])
df['MSZoning'] = df['MSZoning'].fillna(df['MSZoning'].mode()[0])
df['Exterior1st'] = df['Exterior1st'].fillna(df['Exterior1st'].mode()[0])
df['BsmtFinSF2'] = df['BsmtFinSF2'].fillna(df['BsmtFinSF2'].median())
df['TotalBsmtSF'] = df['TotalBsmtSF'].fillna(df['TotalBsmtSF'].median())

# Remove duplicates
df = df.drop_duplicates()

# Drop Id column
df = df.drop('Id', axis=1)

# One-hot encoding
df_encoded = pd.get_dummies(df, columns=['MSZoning', 'LotConfig', 'BldgType', 'Exterior1st'], drop_first=True)

# Verify
print("Final shape:", df_encoded.shape)
df_encoded.head()
```

**# Task 1**

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_encoded.drop('SalePrice', axis=1)
print(x)
```

```python
y = df_encoded['SalePrice']
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print("x_train shape:", x_train.shape)
```

```python
print("x_test shape:", x_test.shape)
```

```python
print("y_train shape:", y_train.shape)
```

```python
print("y_test shape:", y_test.shape)
```

# **Task 2**

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
print(model)
```

```python
model.fit(x_train, y_train)
```

# **Task 3**

```python
y_pred = model.predict(x_test)
```

```python
y_pred[:10]
```

# **Task 4**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
mae = mean_absolute_error(y_test, y_pred)
print(mae)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(mse)
```

```python
rmse = np.sqrt(mse)
print(rmse)
```

```python
r2 = r2_score(y_test, y_pred)
print(r2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R² Score:", r2)
```
