```python
# Topic 52  NumPy Indexing, Slicing & Strides
import numpy as np
arr=np.array([[10,20,30],
              [40,50,60],
              [70,80,90]])
print(arr)
```

```python
print(arr[0,1]) # 20 Row 0 column 1
print(arr[2,2]) # 90 Row 2 column 2
```

```python
print(arr[-1,-2]) # 80 Last Row 2nd Last Column
print(arr[-3,-3]) # 10 First Row First Column
```

```python
# General Form: arr[start:stop:step]
print(arr[0:2,1:3]) # Row 0-1 Column 1-2
```

```python
print(arr[::2,::2]) # Every 2nd Row and 2nd column
```

```python
slice_arr=arr[0,:2]
slice_arr[0]=999    # View Not a copy
```

```python
print(arr[0])  # original changes
```

```python
copy_arr=arr[1].copy() # Independent copy
copy_arr[0]=555
```

```python
print(arr[1])
```

```python
# Topic 53 NumPy Random Numbers, Reproducibility & Performance Tips
# Simulat 100k coin flips
import numpy as np
# np.random.seed(42)
```

```python
rng=np.random.default_rng()
flips=rng.integers(0,2,100000)
prob_heads=np.mean(flips)
print("Estimated Probibility of Head: ", prob_heads)
```

```python
# Simulation with seed
rng=np.random.default_rng(seed=43) # Give us same numbers
flips=rng.integers(0,2,100000)
prob_heads=np.mean(flips)
print("Estimated Probibility of Head: ",prob_heads)
```

```python
# Topic 56 Pandas Indexing & Selection - loc & iloc

data={
    'Name':['Alice', 'Bob', 'Charlie','David'],
    'Age':[25,40,30,35],
    'City':['New York','Paris','London','Berlin'],
    'Salary':[40000,50000,60000,45000]

}
df=pd.DataFrame(data)
print("Original DataFrame:\n", data)
```

```python
# 1. label base selection (.loc)
subset_loc=df.loc[1:3,['Name','Salary']]
print("Subset With loc:\n",subset_loc)
```

```python
from re import sub
# Integer Base Selection (.iloc)
subset_iloc=df.iloc[0:2,1:3]
print("Subset With iloc:\n",subset_iloc) # time 4:20
```

```python
# Boolean filtering
high_salary=df[df['Salary']>40000]
print("High Salary:\n",high_salary)
```

```python
# safe assignment using (.loc)
df.loc[df['Age'] <40,'Salary']=60000
print("Low Age Salary Change:\n", df.loc[df['Age']<40,'Salary'])
```
