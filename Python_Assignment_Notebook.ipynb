### Step 1: Conditional Statements (Tax Calculator)

```python
income = float(input("Enter annual income: "))

if income <= 400000:
    tax_rate = 0
    tax_amount = 0
elif income <= 800000:
    tax_rate = 5
    tax_amount = income * 0.05
elif income <= 1200000:
    tax_rate = 10
    tax_amount = income * 0.10
else:
    tax_rate = 20
    tax_amount = income * 0.20

print(f"Tax Rate: {tax_rate}%")
print(f"Total Tax: {tax_amount}")
```

### Step 2: Lists and Dictionaries (Data Mapping)

```python
products = ["Laptop", "Mouse", "Keyboard", "Monitor"]
product_catalog = {}
start_id = 1001

for item in products:
    product_catalog[start_id] = item
    start_id += 1

print("Product Catalog:", product_catalog)
```

### Step 3: Object-Oriented Programming (Chat System)

```python
class ChatUser:
    def __init__(self, name):
        self.name = name
        self.history = []

    def send_message(self, recipient, text):
        msg = f"[{self.name}]: {text}"
        self.history.append(f"To {recipient.name}: {text}")
        recipient.receive_message(msg)

    def receive_message(self, msg):
        self.history.append(msg)

user1 = ChatUser("Ali")
user2 = ChatUser("Sara")
user1.send_message(user2, "Hello!")
print("Message sent successfully!")
```
