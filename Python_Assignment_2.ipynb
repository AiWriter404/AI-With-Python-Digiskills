```python
import pandas as pd


df = pd.read_csv('/kaggle/input/datasets/muhammadumer7804/house-price-dataset/HousePricePrediction.csv')


df.head(10)
```

```python

print("Shape:", df.shape)
```

```python
print("\nData Types:\n")
print(df.dtypes)
```

```python
print("\nSummary Statistics:\n")
df.describe()
```

```python
print("Missing Values:\n")
print(df.isnull().sum())
```

```python
df = df.dropna(subset=['SalePrice'])

df['MSZoning'] = df['MSZoning'].fillna(df['MSZoning'].mode()[0])

df['Exterior1st'] = df['Exterior1st'].fillna(df['Exterior1st'].mode()[0])

df['BsmtFinSF2'] = df['BsmtFinSF2'].fillna(df['BsmtFinSF2'].median())
df['TotalBsmtSF'] = df['TotalBsmtSF'].fillna(df['TotalBsmtSF'].median())
# Verify
print("New shape:", df.shape)
print("\nMissing values after cleaning:\n")
print(df.isnull().sum())
```

```python
print("Duplicate rows:", df.duplicated().sum())

df = df.drop_duplicates()

# Verify
print("Shape after removing duplicates:", df.shape)
```

```python
# Task 4: Feature Selection

print("Current columns:\n", df.columns.tolist())

df = df.drop('Id', axis=1)

# Verify
print("\nColumns after dropping 'Id':\n", df.columns.tolist())
print("\nNew shape:", df.shape)
```

```python
# Task 5: Data Preprocessing
categorical_cols = df.select_dtypes(include='object').columns
print("Categorical columns:", categorical_cols.tolist())
```

```python
# One-hot encoding apply karein categorical columns pe
df_encoded = pd.get_dummies(df, columns=['MSZoning', 'LotConfig', 'BldgType', 'Exterior1st'], drop_first=True)

# Preview dekhein - final ML-ready dataset
print("Shape after encoding:", df_encoded.shape)
df_encoded.head()
```
