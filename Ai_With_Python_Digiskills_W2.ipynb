```python
# 1. Aik dafa Class (Blueprint) bana li
class Car:
    def __init__(self, brand, model):
        self.brand = brand  # Car ki property
        self.model = model  # Car ki property

    def show_details(self): # Car ka action/function
        print(f"Car: {self.brand} {self.model}")

# 2. Ab is se jitni marzi Objects (Real Cars) banayein
car1 = Car("Toyota", "Corolla")
car2 = Car("Honda", "Civic")
car3 = Car("Suzuki", "Alto")

# 3. Har car apna function khud call karegi
car1.show_details()
car2.show_details()
car3.show_details()
```

```python
# Topic 29 | Core OOP Concepts - Classes, Objects, Methods & Attributes
# 1. CLASS (Naqsha/Blueprint)
class Student:
    
    # __init__ constructor hai jo Attributes ko initialize karta hai
    def __init__(self, name, roll_number):
        # ATTRIBUTES (Khusoosiyat/Data)
        self.name = name              
        self.roll_number = roll_number 

    # METHOD (Action/Function)
    def introduce(self):
        print(f"Assalam-o-Alaikum! Mera naam {self.name} hai aur mera Roll Number {self.roll_number} hai.")

# =======================================================

# 2. OBJECTS (Asli instances jo class se bani hain)
student1 = Student("Umar", 101)
student2 = Student("Ali", 102)

# Attributes ko access karna
print(student1.name) # Output: Umar

# Methods ko call karna (Actions perform karna)
student1.introduce() # Output: Assalam-o-Alaikum! Mera naam Umar hai...
student2.introduce()
```

```python
# Topic 30 | Defining a Class & Creating Objects
# 1. Class ko DEFINE karna
class Smartphone:
    # Constructor jo naye phone ki details (Attributes) set karega
    def __init__(self, brand, model, price):
        self.brand = brand     # Phone ki brand (e.g., Samsung)
        self.model = model     # Phone ka model (e.g., S24)
        self.price = price     # Phone ki keemat (e.g., 150000)

    # Aik simple method (Function) phone ki details dikhane ke liye
    def display_info(self):
        print(f"Phone: {self.brand} {self.model} | Price: Rs.{self.price}")


# =======================================================


# 2. Objects ko CREATE karna (Class se asli phones banana)
phone1 = Smartphone("Apple", "iPhone 15", 250000)
phone2 = Smartphone("Samsung", "Galaxy S24", 220000)


# 3. Objects ko USE karna
# Phone 1 ki details print karna
phone1.display_info()  # Output: Phone: Apple iPhone 15 | Price: Rs.250000

# Phone 2 ki details print karna
phone2.display_info()  # Output: Phone: Samsung Galaxy S24 | Price: Rs.220000
```

```python
# Topic 31 | The __init__() Constructor Method
class Gamer:
    # __init__ constructor define kiya
    def __init__(self, username, game):
        self.username = username  # Player ka naam set ho raha hai
        self.game = game          # Player ki game set ho rahi hai
        print(f"🎮 Naya player '{self.username}' join ho chuka hai!")

    # Ek simple method player ki profile dikhane ke liye
    def show_profile(self):
        print(f"Player: {self.username} | Favourite Game: {self.game}")


# =======================================================


# Jaise hi hum object banayenge, __init__ khud chal parega aur print statement run hogi
player1 = Gamer("Umar_Pro", "PUBG") 
player2 = Gamer("Ali_Rider", "NFS")

# Ab hum doosra method call kar sakte hain
player1.show_profile()
```

```python
# Topic 32 | Instance Variables & Instance Methods
class BankAccount:
    # __init__ ke andar instance variables bante hain
    def __init__(self, account_holder, balance):
        self.holder = account_holder  # Instance Variable 1
        self.balance = balance        # Instance Variable 2

    # Instance Method: Jo balance check karega
    def check_balance(self):
        print(f"{self.holder} ka mojooda balance: Rs.{self.balance}")

    # Instance Method: Jo paise jama (deposit) karega
    def deposit(self, amount):
        self.balance += amount
        print(f"Rs.{amount} jama ho gaye! Naya balance: Rs.{self.balance}")


# =======================================================


# 1. Alag alag objects (instances) banayein
account1 = BankAccount("Umar", 5000)
account2 = BankAccount("Ali", 12000)

# 2. Instance Variables check karein (Dono ka data alag hai)
print(account1.holder)  # Output: Umar
print(account2.holder)  # Output: Ali

# 3. Instance Methods ko call karein
account1.check_balance()  # Output: Umar ka mojooda balance: Rs.5000
account1.deposit(2000)    # Output: Rs.2000 jama ho gaye! Naya balance: Rs.7000

print("---")

# account2 ka balance bilkul alag rahega, us par koi farq nahi para
account2.check_balance()  # Output: Ali ka mojooda balance: Rs.12000
```

```python
# Topic 33 | Accessing Object Attributes & Methods
class Student:
    def __init__(self, name, age):
        self.name = name  # Attribute 1
        self.age = age    # Attribute 2

    # Simple Method
    def display_info(self):
        print(f"Student Name: {self.name}, Age: {self.age}")


# =======================================================


# 1. Object Create Kiya
s1 = Student("Umar", 21)


# 2. Attributes ko Access (Read) karna
print("--- Attributes Accessing ---")
print(s1.name)  # Output: Umar
print(s1.age)   # Output: 21


# 3. Attributes ki Value ko Change (Modify) karna
s1.age = 22     # Umar ki age 21 se badal kar 22 kar di
print(s1.age)   # Output: 22


# 4. Method ko Access (Call) karna
print("--- Method Accessing ---")
s1.display_info()  # Output: Student Name: Umar, Age: 22
```

```python
# Topic 34 | Encapsulation in Python - Public & Private Attributes
class Employee:
    def __init__(self, name, salary):
        self.name = name          # Public Attribute (Sab dekh sakte hain)
        self.__salary = salary    # Private Attribute (Double underscore lagane se chup gaya)

    # Private data ko dekhne ke liye aik Public Method (Getter)
    def show_salary(self):
        print(f"{self.name} ki salary Rs.{self.__salary} hai.")


# =======================================================


emp = Employee("Umar", 50000)

# 1. Public attribute ko access karna (Yeh kaam karega)
print(emp.name)  # Output: Umar
emp.name = "Umar Khan" # Change bhi ho sakta hai

# 2. Private attribute ko direct access karna (Yeh ERROR dega)
# print(emp.__salary)  # AttributeError: 'Employee' object has no attribute '__salary'

# 3. Private data ko sahi tariqe se dekhna (Method ke zariye)
emp.show_salary()  # Output: Umar Khan ki salary Rs.50000 hai.
```

```python
# Topic 35 | Using Getters & Setters in Python
class MobileAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private Attribute

    # 1. GETTER: Balance dekhne ke liye
    def get_balance(self):
        return self.__balance

    # 2. SETTER: Balance ko tameez se badalny ke liye
    def set_balance(self, new_amount):
        # Validation: Balance kabhi minus (-) mein nahi ho sakta
        if new_amount >= 0:
            self.__balance = new_amount
            print("Balance kamyabi se update ho gaya!")
        else:
            print("Error: Balance minus mein nahi ho sakta!")


# =======================================================


user = MobileAccount(500)

# Direct access block hai (Error aayega)
# print(user.__balance) 

# 1. Getter use karke balance check kiya
print("Purana Balance:", user.get_balance()) # Output: Purana Balance: 500

# 2. Setter use karke sahi value set ki
user.set_balance(1000) 
print("Naya Balance:", user.get_balance())   # Output: Naya Balance: 1000

# 3. Setter mein ghalat value de kar check kiya (Validation Test)
user.set_balance(-200)                       # Output: Error: Balance minus mein nahi ho sakta!
```

```python
# Topic 36 | Inheritance in Python - Parent & Child Classes
# 1. PARENT CLASS (Base Class)
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} khana kha raha hai.")

# =======================================================

# 2. CHILD CLASS (Derived Class)
# Dog ke aage brackets () mein Animal likhne se Inheritance hoti hai
class Dog(Animal):
    
    # Child class ka apna naya method
    def bark(self):
        print(f"{self.name} bhonk raha hai! Woof Woof!")


# =======================================================


# 3. Child Class ka Object banana
my_dog = Dog("Tommy")

# Child class ka object Parent ke methods ko bhi use kar sakta hai
my_dog.eat()  # Output: Tommy khana kha raha hai. (Inherited from Animal)

# Child class apne unique methods ko bhi use kar sakti hai
my_dog.bark() # Output: Tommy bhonk raha hai! Woof Woof!
```
