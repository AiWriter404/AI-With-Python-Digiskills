```python
# 1. Aik dafa Class (Blueprint) bana li
class Car:
    def __init__(self, brand, model):
        self.brand = brand  # Car ki property
        self.model = model  # Car ki property

    def show_details(self): # Car ka action/function
        print(f"Car: {self.brand} {self.model}")

# 2. Ab is se jitni marzi Objects (Real Cars) banayein
car1 = Car("Toyota", "Corolla")
car2 = Car("Honda", "Civic")
car3 = Car("Suzuki", "Alto")

# 3. Har car apna function khud call karegi
car1.show_details()
car2.show_details()
car3.show_details()
```

```python
# Topic 29 | Core OOP Concepts - Classes, Objects, Methods & Attributes
# 1. CLASS (Naqsha/Blueprint)
class Student:
    
    # __init__ constructor hai jo Attributes ko initialize karta hai
    def __init__(self, name, roll_number):
        # ATTRIBUTES (Khusoosiyat/Data)
        self.name = name              
        self.roll_number = roll_number 

    # METHOD (Action/Function)
    def introduce(self):
        print(f"Assalam-o-Alaikum! Mera naam {self.name} hai aur mera Roll Number {self.roll_number} hai.")

# =======================================================

# 2. OBJECTS (Asli instances jo class se bani hain)
student1 = Student("Umar", 101)
student2 = Student("Ali", 102)

# Attributes ko access karna
print(student1.name) # Output: Umar

# Methods ko call karna (Actions perform karna)
student1.introduce() # Output: Assalam-o-Alaikum! Mera naam Umar hai...
student2.introduce()
```

```python
# Topic 30 | Defining a Class & Creating Objects
# 1. Class ko DEFINE karna
class Smartphone:
    # Constructor jo naye phone ki details (Attributes) set karega
    def __init__(self, brand, model, price):
        self.brand = brand     # Phone ki brand (e.g., Samsung)
        self.model = model     # Phone ka model (e.g., S24)
        self.price = price     # Phone ki keemat (e.g., 150000)

    # Aik simple method (Function) phone ki details dikhane ke liye
    def display_info(self):
        print(f"Phone: {self.brand} {self.model} | Price: Rs.{self.price}")


# =======================================================


# 2. Objects ko CREATE karna (Class se asli phones banana)
phone1 = Smartphone("Apple", "iPhone 15", 250000)
phone2 = Smartphone("Samsung", "Galaxy S24", 220000)


# 3. Objects ko USE karna
# Phone 1 ki details print karna
phone1.display_info()  # Output: Phone: Apple iPhone 15 | Price: Rs.250000

# Phone 2 ki details print karna
phone2.display_info()  # Output: Phone: Samsung Galaxy S24 | Price: Rs.220000
```

```python
# Topic 32 | Instance Variables & Instance Methods
class BankAccount:
    # __init__ ke andar instance variables bante hain
    def __init__(self, account_holder, balance):
        self.holder = account_holder  # Instance Variable 1
        self.balance = balance        # Instance Variable 2

    # Instance Method: Jo balance check karega
    def check_balance(self):
        print(f"{self.holder} ka mojooda balance: Rs.{self.balance}")

    # Instance Method: Jo paise jama (deposit) karega
    def deposit(self, amount):
        self.balance += amount
        print(f"Rs.{amount} jama ho gaye! Naya balance: Rs.{self.balance}")


# =======================================================


# 1. Alag alag objects (instances) banayein
account1 = BankAccount("Umar", 5000)
account2 = BankAccount("Ali", 12000)

# 2. Instance Variables check karein (Dono ka data alag hai)
print(account1.holder)  # Output: Umar
print(account2.holder)  # Output: Ali

# 3. Instance Methods ko call karein
account1.check_balance()  # Output: Umar ka mojooda balance: Rs.5000
account1.deposit(2000)    # Output: Rs.2000 jama ho gaye! Naya balance: Rs.7000

print("---")

# account2 ka balance bilkul alag rahega, us par koi farq nahi para
account2.check_balance()  # Output: Ali ka mojooda balance: Rs.12000
```

```python
# Topic 33 | Accessing Object Attributes & Methods
class Student:
    def __init__(self, name, age):
        self.name = name  # Attribute 1
        self.age = age    # Attribute 2

    # Simple Method
    def display_info(self):
        print(f"Student Name: {self.name}, Age: {self.age}")


# =======================================================


# 1. Object Create Kiya
s1 = Student("Umar", 21)


# 2. Attributes ko Access (Read) karna
print("--- Attributes Accessing ---")
print(s1.name)  # Output: Umar
print(s1.age)   # Output: 21


# 3. Attributes ki Value ko Change (Modify) karna
s1.age = 22     # Umar ki age 21 se badal kar 22 kar di
print(s1.age)   # Output: 22


# 4. Method ko Access (Call) karna
print("--- Method Accessing ---")
s1.display_info()  # Output: Student Name: Umar, Age: 22
```

```python
# Topic 34 | Encapsulation in Python - Public & Private Attributes
class Employee:
    def __init__(self, name, salary):
        self.name = name          # Public Attribute (Sab dekh sakte hain)
        self.__salary = salary    # Private Attribute (Double underscore lagane se chup gaya)

    # Private data ko dekhne ke liye aik Public Method (Getter)
    def show_salary(self):
        print(f"{self.name} ki salary Rs.{self.__salary} hai.")


# =======================================================


emp = Employee("Umar", 50000)

# 1. Public attribute ko access karna (Yeh kaam karega)
print(emp.name)  # Output: Umar
emp.name = "Umar Khan" # Change bhi ho sakta hai

# 2. Private attribute ko direct access karna (Yeh ERROR dega)
# print(emp.__salary)  # AttributeError: 'Employee' object has no attribute '__salary'

# 3. Private data ko sahi tariqe se dekhna (Method ke zariye)
emp.show_salary()  # Output: Umar Khan ki salary Rs.50000 hai.
```

```python
# Topic 35 | Using Getters & Setters in Python
class MobileAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private Attribute

    # 1. GETTER: Balance dekhne ke liye
    def get_balance(self):
        return self.__balance

    # 2. SETTER: Balance ko tameez se badalny ke liye
    def set_balance(self, new_amount):
        # Validation: Balance kabhi minus (-) mein nahi ho sakta
        if new_amount >= 0:
            self.__balance = new_amount
            print("Balance kamyabi se update ho gaya!")
        else:
            print("Error: Balance minus mein nahi ho sakta!")


# =======================================================


user = MobileAccount(500)

# Direct access block hai (Error aayega)
# print(user.__balance) 

# 1. Getter use karke balance check kiya
print("Purana Balance:", user.get_balance()) # Output: Purana Balance: 500

# 2. Setter use karke sahi value set ki
user.set_balance(1000) 
print("Naya Balance:", user.get_balance())   # Output: Naya Balance: 1000

# 3. Setter mein ghalat value de kar check kiya (Validation Test)
user.set_balance(-200)                       # Output: Error: Balance minus mein nahi ho sakta!
```

```python
# Topic 36 | Inheritance in Python - Parent & Child Classes
# 1. PARENT CLASS (Base Class)
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} khana kha raha hai.")

# =======================================================

# 2. CHILD CLASS (Derived Class)
# Dog ke aage brackets () mein Animal likhne se Inheritance hoti hai
class Dog(Animal):
    
    # Child class ka apna naya method
    def bark(self):
        print(f"{self.name} bhonk raha hai! Woof Woof!")


# =======================================================


# 3. Child Class ka Object banana
my_dog = Dog("Tommy")

# Child class ka object Parent ke methods ko bhi use kar sakta hai
my_dog.eat()  # Output: Tommy khana kha raha hai. (Inherited from Animal)

# Child class apne unique methods ko bhi use kar sakti hai
my_dog.bark() # Output: Tommy bhonk raha hai! Woof Woof!
```

```python
# Topic 37 | Method Overriding
# 1. PARENT CLASS
class Animal:
    def make_sound(self):
        print("Yeh animal koi aawaz nikal raha hai.")

# =======================================================

# 2. CHILD CLASS 1 (Overriding bina kisi tabdeeli ke)
class Dog(Animal):
    # Method Overriding: Same naam ka function dobara likha
    def make_sound(self):
        print("Woof! Woof!")

# 3. CHILD CLASS 2 (Overriding apni aawaz ke sath)
class Cat(Animal):
    # Method Overriding: Same naam ka function dobara likha
    def make_sound(self):
        print("Meow! Meow!")


# =======================================================


# Objects banana
generic_animal = Animal()
puppy = Dog()
kitty = Cat()

# Alag alag outputs check karein
generic_animal.make_sound() # Output: Yeh animal koi aawaz nikal raha hai.
puppy.make_sound()          # Output: Woof! Woof!  (Parent ka method dab gaya)
kitty.make_sound()          # Output: Meow! Meow!  (Parent ka method dab gaya)
```

```python
# Topic 38 | Polymorphism in Python - Overriding & Overloading
class Pakistan:
    def capital(self):
        print("Islamabad")

class UAE:
    def capital(self):
        print("Abu Dhabi")

# Aik hi function jo kisi bhi country ka object le kar uska capital bataye
def print_capital(country_object):
    country_object.capital() # Naam same hai, kaam object par depend karta hai

pak = Pakistan()
uae = UAE()

print_capital(pak)  # Output: Islamabad
print_capital(uae)  # Output: Abu Dhabi
```

```python
class Calculator:
    # Default values None set kar dein taake arguments change ho sakein
    def add(self, a, b, c = None):
        if c is not None:
            return a + b + c
        else:
            return a + b

calc = Calculator()

# Situation 1: Do numbers ko plus karna
print(calc.add(5, 10))     # Output: 15

# Situation 2: Teen numbers ko plus karna (Same function, different input)
print(calc.add(5, 10, 20))  # Output: 35
```

```python
# Topic 39 | Python Special Methods
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    # 1. __str__ Method: Object ko readable text mein badalna
    def __str__(self):
        return f"Kitab ka Naam: '{self.title}'"

    # 2. __len__ Method: Object ki length batana
    def __len__(self):
        return self.pages


# =======================================================


# Object create kiya
my_book = Book("Python AI Programming", 350)

# Pehle agar hum direct object print karte to ganda sa memory address aata.
# Lekin __str__ ki wajah se ab saaf text aayega:
print(my_book)  # Output: Kitab ka Naam: 'Python AI Programming'

# __len__ ki wajah se hum custom object ki length maloom kar sakte hain:
print("Total Pages:", len(my_book))  # Output: Total Pages: 350
```

```python
# Topic 43 | Handling Exceptions During File Operations
file_name = "asli_file.txt" # Yeh file computer mein nahi hai

try:
    # 1. try block mein woh code likhte hain jahan error aa sakta hai
    with open(file_name, "r") as file:
        content = file.read()
        print(content)

except FileNotFoundError:
    # 2. Agar file na mile to yeh block chalega aur program crash nahi hoga
    print(f"❌ Error: '{file_name}' naam ki koi file nahi mili. Pehle file banayein!")

except PermissionError:
    # 3. Agar file open karne ki ijazat na ho (Protected file)
    print("❌ Error: Aap ke paas is file ko open karne ki permission nahi hai.")

except Exception as e:
    # 4. Kisi bhi aur qism ka anjaan error handle karne ke liye
    print(f"❌ Koi unexpected error aaya hai: {e}")

# Try-except ki wajah se yeh neechay wali line lazmi chalegi
print("\nProgram abhi bhi chal raha hai aur crash nahi hua!")
```

```python
# Topic 44 | Introduction to Exceptions in Python
num1 = 10
num2 = 0

try:
    result = num1 / num2
except ZeroDivisionError:
    print("❌ Error: Zero se divide nahi kiya ja sakta!")
```

```python
# Topic 45 | Using try, except, else & finally Blocks
try:
    # 1. User se inputs lena (Yahan error aa sakta hai agar user 0 ya string de)
    #num = int(input("Koi bhi number likhein: "))
    num =4
    result = 100 / num

except ZeroDivisionError:
    # 2. Agar user ne 0 likha to yeh chalega
    print("❌ Error: Aap 100 ko zero (0) se divide nahi kar sakte!")

except ValueError:
    # 3. Agar user ne number ke bajaye text likh diya to yeh chalega
    print("❌ Error: Meharbani karke sirf ginti (integer) likhein!")

else:
    # 4. Agar koi error nahi aaya to yeh chalega
    print(f"✅ Kamyabi! Aapka result hai: {result}")

finally:
    # 5. Yeh har haal mein chalega hi chalega
    print("🧹 Clean-up: Division ka process mukammal ho gaya.")
```

```python
# Topic 46 | Raising Exceptions in Python
def check_voting_eligibility(age):
    # 1. Agar age minus mein ho to ValueError raise karein
    if age < 0:
        raise ValueError("Umar minus (-) mein nahi ho sakti!")
    
    # 2. Agar age 18 se kam ho to custom message ke sath Exception raise karein
    if age < 18:
        raise Exception("Aap voting ke liye eligible nahi hain kyunke aapki umar 18 se kam hai.")
    
    print("✅ Mubarak ho! Aap vote daal sakte hain.")

# =======================================================

# 3. Code ko try-except mein chala kar check karte hain
try:
    check_voting_eligibility(15)  # Umar 18 se kam de kar check kiya
except Exception as error:
    print(f"❌ Pakra gaya error: {error}")
```

```python
# Topic 47 | Creating Custom Exceptions in Python
# 1. Custom Exception Class banayi (Jo Exception class se inherit ho rahi hai)
class InvalidUsernameError(Exception):
    """Username ke rules tootne par yeh error aayega."""
    pass 


# =======================================================


# 2. Ek function jo is custom exception ko use karega
def register_user(username):
    if " " in username:
        # Custom Exception ko raise kiya
        raise InvalidUsernameError("Username mein space (' ') lagana mana hai!")
    
    print(f"✅ User '{username}' kamyabi se register ho gaya!")


# =======================================================


# 3. Code ko try-except block mein chala kar check karte hain
try:
    user_input = "Umar Pro" # Is mein space hai, to error aana chahiye
    register_user(user_input)

except InvalidUsernameError as error:
    # Hamari apni custom exception yahan pakri gayi
    print(f"❌ Custom Error Pakra Gaya: {error}")
```
