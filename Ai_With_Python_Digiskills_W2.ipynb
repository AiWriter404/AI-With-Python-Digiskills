```python
# Topic 28 | Understanding the Need For OOP
# Car 1 ka data
car1_brand = "Toyota"
car1_model = "Corolla"

# Car 2 ka data
car2_brand = "Honda"
car2_model = "Civic"

# Car ki details print karne ka function
def print_car_details(brand, model):
    print(f"Car: {brand} {model}")

print_car_details(car1_brand, car1_model)
print_car_details(car2_brand, car2_model)
```

```python
# 1. Aik dafa Class (Blueprint) bana li
class Car:
    def __init__(self, brand, model):
        self.brand = brand  # Car ki property
        self.model = model  # Car ki property

    def show_details(self): # Car ka action/function
        print(f"Car: {self.brand} {self.model}")

# 2. Ab is se jitni marzi Objects (Real Cars) banayein
car1 = Car("Toyota", "Corolla")
car2 = Car("Honda", "Civic")
car3 = Car("Suzuki", "Alto")

# 3. Har car apna function khud call karegi
car1.show_details()
car2.show_details()
car3.show_details()
```

```python
# Topic 29 | Core OOP Concepts - Classes, Objects, Methods & Attributes
# 1. CLASS (Naqsha/Blueprint)
class Student:
    
    # __init__ constructor hai jo Attributes ko initialize karta hai
    def __init__(self, name, roll_number):
        # ATTRIBUTES (Khusoosiyat/Data)
        self.name = name              
        self.roll_number = roll_number 

    # METHOD (Action/Function)
    def introduce(self):
        print(f"Assalam-o-Alaikum! Mera naam {self.name} hai aur mera Roll Number {self.roll_number} hai.")

# =======================================================

# 2. OBJECTS (Asli instances jo class se bani hain)
student1 = Student("Umar", 101)
student2 = Student("Ali", 102)

# Attributes ko access karna
print(student1.name) # Output: Umar

# Methods ko call karna (Actions perform karna)
student1.introduce() # Output: Assalam-o-Alaikum! Mera naam Umar hai...
student2.introduce()
```

```python
# Topic 30 | Defining a Class & Creating Objects
# 1. Class ko DEFINE karna
class Smartphone:
    # Constructor jo naye phone ki details (Attributes) set karega
    def __init__(self, brand, model, price):
        self.brand = brand     # Phone ki brand (e.g., Samsung)
        self.model = model     # Phone ka model (e.g., S24)
        self.price = price     # Phone ki keemat (e.g., 150000)

    # Aik simple method (Function) phone ki details dikhane ke liye
    def display_info(self):
        print(f"Phone: {self.brand} {self.model} | Price: Rs.{self.price}")


# =======================================================


# 2. Objects ko CREATE karna (Class se asli phones banana)
phone1 = Smartphone("Apple", "iPhone 15", 250000)
phone2 = Smartphone("Samsung", "Galaxy S24", 220000)


# 3. Objects ko USE karna
# Phone 1 ki details print karna
phone1.display_info()  # Output: Phone: Apple iPhone 15 | Price: Rs.250000

# Phone 2 ki details print karna
phone2.display_info()  # Output: Phone: Samsung Galaxy S24 | Price: Rs.220000
```
